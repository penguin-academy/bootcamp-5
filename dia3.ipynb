### Ejercicio

Escribir un programa que pregunte al usuario por la contraseña hasta que introduzca la contraseña correcta.

```python
# Solución de Dani

intento = 4
contrasenha = None

while contrasenha != "Spiderman" and intento > 0:
    contrasenha = input("Ingrese tu contraseña: ")

    if contrasenha == "Spiderman":
        print("Contraseña correcta. Bienvenido.")
    else:
        intento = intento - 1
        print("Contraseña incorrecta. ¡Vuelva a intentarlo!", "Te sobra", intento, "intentos.")
    if intento == 0:
        print("Sesión caducada.")
```

```python
# Solución de Maura

intentos=3
contraseña_ingresada= None
aun_le_sobran_intentos=True
la_contraseña_es_incorrecta = True

while la_contraseña_es_incorrecta and aun_le_sobran_intentos:
    contraseña_ingresada = input ("Escriba su contraseña")
    la_contraseña_es_incorrecta = contraseña_ingresada != "penguin"

    intentos = intentos - 1
    aun_le_sobran_intentos = intentos > 0

    if la_contraseña_es_incorrecta:
        print ("La contraseña es incorrecta.Le sobran ", intentos, "intentos.")
    if contraseña_ingresada == "penguin":
        print ("contraseña correcta")
```

### Ejercicio
- Imprimir el listado de productos.
- Ejemplo: 
    - Producto #1: Leche
    - Producto #2: Pan

```python
# Solución de Tami

inventario = ["leche", "pan", "azucar", "huevo", "queso"]

cantidad_de_productos_del_inventario = len(inventario)

for posicion_del_producto_en_el_inventario in range(cantidad_de_productos_del_inventario):
    print("Producto #" + str(posicion_del_producto_en_el_inventario + 1) + ": " + inventario[posicion_del_producto_en_el_inventario])
```

```python
inventario = ["leche", "pan", "azucar", "huevo", "queso"]

cantidad = len(inventario)
for producto in range(cantidad):
    print("Producto Nro.", producto + 1 , inventario[producto])
```

### Ejercicio

Escribir un programa en el que se pregunte al usuario por una frase y una letra, y muestre por pantalla el número de veces que aparece la letra en la frase.

```python
# Solución de Luis

frase = input("Por favor, ingrese una palabra ")
letra = input("Escoja la letra que desee contar ")
contador = 0

cantidad_de_las_letras_dentro_de_la_frase = len(frase)

for indice_de_las_letras_dentro_de_la_frase in range(cantidad_de_las_letras_dentro_de_la_frase):
    if frase[indice_de_las_letras_dentro_de_la_frase] in letra:
        contador = contador + 1

print(contador)
```

```python
# Solución de Rafa

frase_usuario = input("Escriba una frase, debe estar solamente en minusculas: ")
letra_deseada = input("Escriba la letra que usted desea buscar, solamente en minusculas: ")

letra_repetida = []

numero_repeticiones = len(letra_repetida)

for letra in frase_usuario:
    if letra == letra_deseada:
        letra_repetida.append(letra)

print("En esta frase hay", len(letra_repetida), letra_deseada)
```

```python
# Solución con count

frase_usuario = input("Escriba una frase, debe estar solamente en minusculas: ")
letra_deseada = input("Escriba la letra que usted desea buscar, solamente en minusculas: ")

print(frase_usuario.count(letra_deseada))
```
