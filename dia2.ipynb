## Operadores lógicos

```python
# Operador and

es_alergica = True
come_mani = False

print(es_alergica and come_mani)

mayor_edad = True
colaboro_vaquita = False

print(mayor_edad and colaboro_vaquita)

# Operador or

es_alergica = True
come_mani = False

print(es_alergica or come_mani)

# Operador not

print(not es_alergica)
```

## Operadores de comparación

```python
# Lista de precios

manzanas = 8800
naranjas = 6000
peras = 7500
manzanas_verdes = 8800

# Comparamos con operadores
# Operador igual que

print(manzanas == manzanas_verdes)
print(naranjas == peras)

# Operador distinto de
print(manzanas != manzanas_verdes)
print(peras != manzanas_verdes)

# Operador mayor que
print(manzanas_verdes > peras)
print(naranjas > manzanas)

# Operador menor que
print(naranjas < peras)
print(manzanas_verdes < manzanas)

# Operador >= mayor o igual que
print(naranjas >= manzanas)
print(manzanas >= manzanas_verdes)

# Operador <= menor o igual que
print(naranjas <= manzanas_verdes)
print(peras <= manzanas)
```

### Ejercicio: Condicionales

1. Si tenés menos de 18, no podés comprar.
2. Si tenés entre 18 y 21, podés tomar pero no comprar.
3. Si tenés más de 21, hacé lo que queres kp.


```python
# Solución de Maura

edad = int(input("Escriba su edad, por favor: "))

if edad < 18:
    print("No puede comprar ni tomar.")
elif edad == 18 or edad < 21:
    print("Puede tomar, pero no comprar.")
else:
    print("Hace lo que quieras.")
```

```python
# Solución de Martín

edad = int(input("Por favor, introduzca su edad: "))
fondos = int(input("Introduzca el monto de dinero que dispone: "))
faltante = str(20000 - fondos)

if edad < 18:
    print("No podes comprar ni tomar.")
elif edad >= 21 and fondos < 20000:
    print("Se requiere un monto de 20mil guaranies, te falta " + faltante)
elif edad >= 21 and fondos >= 20000:
    print("Que te sirvo kp?")
else:
    print("Prohibida la venta a menores de 21")
```

```python
# Solución de Edu

edad = int(input("Por favor, introduzca su edad: "))
pilsen = 10000
brahma = 15000
es_mayor_de_edad = edad >= 18
puede_tomar_no_comprar = es_mayor_de_edad and edad <= 21

if not es_mayor_de_edad:
    print("No puede tomar ni comprar alcohol.")
elif puede_tomar_no_comprar:
    print("Puede tomar, no comprar alcohol.")
else:
    print("Puede tomar y comprar alcohol.")
    plata = int(input("Cuanta plata tenes?"))
    plata_suficiente = plata >= 20000

    if plata_suficiente:
        print("Te alcanza el dinero para comprar alcohol.")

        compra = input("Queres comprar Pilsen o Brahma?")
        compro_Pilsen = compra == "PILSEN"
        compro_Brahma = compra == "BRAHMA"

        if compro_Pilsen:
            print("Te alcanza para", plata // pilsen, "botellas. Vuelto:", plata % pilsen)
        else:
            print("Te alcanza para", plata // brahma, "botellas. Vuelto:", plata % brahma)
    else:
        print("No te alcanza el dinero para comprar alcohol.")
```

```python
# Solución de Fer

edad = int(input("Escriba su edad, por favor: "))
es_mayor_de_edad = edad >= 18 # True / False
puede_comprar = edad >= 21 # True / False

if not es_mayor_de_edad:
    print("No podes comprar, ni tomar.")
elif es_mayor_de_edad and not puede_comprar:
    print("No podes comprar, pero si tomar.")
else:
    plata_del_cliente = int(input("Escriba su monto, por favor: "))
    tiene_plata_suficiente = plata_del_cliente >= 10000

    if tiene_plata_suficiente:
        print("Podes comprar, tomar, y tu dinero es suficiente.")
    else:
        print("Podes comprar, tomar, pero tu dinero no es suficiente.")
```

## Listas

```python
lista_frutas = ["naranjas", "peras", "piñas", "kiwis", "sandias"]

# Diferentes tipos de datos en lista
lista_bool = [True, False, True, False]
lista_numeros = [1.99, 2, 3, 4.5, 5]
lista_variada = ["peras", 10, False]

# Acceso a los elementos de la lista
print(lista_frutas[0])
print(lista_frutas[2])

# Recorrer la lista hacia atrás
print(lista_frutas[-1])

# Contar elementos en lista
print(len(lista_frutas))

# Añadir elementos a la lista
lista_frutas.append("manzanas")
print(lista_frutas)

# Añadir elemento en un índice específico
lista_frutas.insert(2, "toronjas")
print(lista_frutas)

# Eliminar elemento específico por valor
lista_frutas.remove("sandias")
print(lista_frutas)

# Eliminar elemento por su índice
lista_frutas.pop(4)
print(lista_frutas)
```
